```python
from traitlets.config.loader import PyFileConfigLoader
import requests
from bs4 import BeautifulSoup
import pandas as pd


new_url = 'https://movies.yahoo.com.tw/movie_thisweek.html'
url = 'https://movies.yahoo.com.tw/movie_intheaters.html'
coming_url = 'https://movies.yahoo.com.tw/movie_comingsoon.html'
page = 3

def take_info(url):

  movie_info = {
    "中文名稱": [],
    "英文名稱": [],
    "上映時間": [],
    "期待度": []
  }

  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html5lib")

  for i in soup.find_all("div", class_="release_info_text"):
    ch_name = i.find("div", class_="release_movie_name").a.text.replace("\n", "").replace(" ","")
    en_name = i.div.div.text.replace("\n", "").replace(" ","")
    date = i.find("div", class_="release_movie_time").text.replace("\n", "").replace(" ","")[5:]
    level = i.find("div", class_="leveltext").span.text
    movie_info["中文名稱"].append(ch_name)
    movie_info["英文名稱"].append(en_name)
    movie_info["上映時間"].append(date)
    movie_info["期待度"].append(level)

  try:
    for i in range(2,page+1):
      link = url + "?page=" + f"{i}"
      new_response = requests.get(link)
      data = BeautifulSoup(new_response.text, "html5lib")

      for j in data.find_all("div", class_="release_info_text"):
        ch_name = j.find("div", class_="release_movie_name").a.text.replace("\n", "").replace(" ","")
        en_name = j.div.div.text.replace("\n", "").replace(" ","")
        date = j.find("div", class_="release_movie_time").text.replace("\n", "").replace(" ","")[5:]
        level = j.find("div", class_="leveltext").span.text
        movie_info["中文名稱"].append(ch_name)
        movie_info["英文名稱"].append(en_name)
        movie_info["上映時間"].append(date)
        movie_info["期待度"].append(level)

  except:
    print(f"{url} : have page fault")

  finally:
    print(f"{url} : Done")
    return(movie_info)

df_url = pd.DataFrame(take_info(url))
df_new = pd.DataFrame(take_info(new_url))
df_coming = pd.DataFrame(take_info(coming_url))
piece = [df_url, df_new, df_coming]
pd.concat(piece, axis = 1)
```
