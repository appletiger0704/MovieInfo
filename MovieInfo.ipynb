```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://movies.yahoo.com.tw/movie_thisweek.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html5lib")

movie_info = {
    "中文名稱": [],
    "英文名稱": [],
    "上映時間": [],
    "期待度": []
    }

for i in soup.find_all("div", class_="release_movie_name"):
  ch_name = i.a.text.strip()
  movie_info["中文名稱"].append(ch_name)
  en_name = i.div.text.strip()
  movie_info['英文名稱'].append(en_name)

for i in soup.find_all("div", class_="release_movie_time"):
  date = i.text.replace("\n","").replace(" ","")[5:]
  movie_info["上映時間"].append(date)

for i in soup.find_all("dl", class_="levelbox"):
  level = i.dt.find("div", class_="leveltext").span.text.replace("\n","").replace(" ","")
  movie_info['期待度'].append(level)

df = pd.DataFrame(movie_info)
df
```

```python
# 即將上映


coming_soon_url = 'https://movies.yahoo.com.tw/movie_comingsoon.html'
coming_soon_response = requests.get(coming_soon_url)
coming_soon_soup = BeautifulSoup(coming_soon_response.text, "html5lib")

coming_soon_info = {
    "中文名稱": [],
    "英文名稱": [],
    "上映時間": [],
    "期待度": []
}
for i in coming_soon_soup.find_all("div", class_="release_info_text"):
  ch_name = i.div.a.text.replace("\n", "").strip()
  coming_soon_info["中文名稱"].append(ch_name)
  en_name = i.div.div.a.text.replace("\n", "").strip()
  coming_soon_info["英文名稱"].append(en_name)
  date = i.find("div", class_="release_movie_time").text.replace("\n", "").replace(" ","")[5:]
  coming_soon_info["上映時間"].append(date)
  level = i.find("div", class_="leveltext").span.text.replace("\n", "").replace(" ","")
  coming_soon_info["期待度"].append(level)

coming_soon_new_soup = coming_soon_soup.find_all("div", class_="page_numbox")[0]
link_tag = coming_soon_new_soup.find_all("a")[:2]

for i in link_tag:
  link = i["href"]
  new_response = requests.get(link)
  new_soup = BeautifulSoup(new_response.text, "html5lib")
  for j in new_soup.find_all("div", class_="release_info_text"):
    ch_name = j.div.a.text.replace("\n", "").strip()
    # print(ch_name)
    coming_soon_info["中文名稱"].append(ch_name)
    en_name = j.div.div.a.text.replace("\n", "").strip()
    # print(en_name)
    coming_soon_info["英文名稱"].append(en_name)
    date = j.find("div", class_="release_movie_time").text.replace("\n", "").replace(" ","")[5:]
    coming_soon_info["上映時間"].append(date)
    # print(date)
    try:
      level = j.find("div", class_="leveltext").span.text
    except:
      level = "none_data"
    finally:
      coming_soon_info["期待度"].append(level)
    # print(level)

df = pd.DataFrame(coming_soon_info)
df
```
